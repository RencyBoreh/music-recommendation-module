```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
music_data=pd.read_csv('music data.csv')
music_data
```

```python

music_data.shape
```

```python
#detailed info of the columns
music_data.describe()
```

```python
#a 2D array
music_data.values
```

```python
#split the data into 2 sets(input and output)
input=music_data.drop(columns=['Genre'])
input
```

```python
output=music_data['Genre']
output
```

```python
#create a model

model= DecisionTreeClassifier()
#train the model
model.fit(input.values,output)
#make prediction
prediction=model.predict([[5,0],[14,1]])#takes 2D array
prediction
```

```python
#measuring the accuracy
#1.split the data into 2(training(80%),testing(20%))
from sklearn.model_selection import train_test_split#with this function we can easily split our data into 2 sets
from sklearn.metrics import accuracy_score
input_train,input_test,output_train,output_test=train_test_split(input,output,test_size=0.2)#This function returns a tuple so we unpacked it
model.fit(input_train,output_train)
predict=model.predict(input_test)#this input_test contains values for tessting
score= accuracy_score(output_test,predict)#compare
score
```
